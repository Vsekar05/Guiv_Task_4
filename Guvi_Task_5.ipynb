# Creating adopted user column based on visited column

```python
final_data["time_stamp"]=pd.to_datetime(final_data["time_stamp"])
```

```python
Grouped_Data=final_data.groupby(['user_id',pd.Grouper(key='time_stamp', freq='W-Mon')])["visited"].sum().reset_index().sort_values("time_stamp")
```

```python
Grouped_Data
```

```python
Grouped_Data["visited"].unique()
```

```python
adopted_user=[]
for i in range(1,38724):
  adopted_user.append(str(i))
```

```python
Grouped_Data["adopted_user"]=adopted_user
```

```python
for i in range(1,38722):
  if (Grouped_Data["visited"][i]>=3):
    Grouped_Data["adopted_user"][i]=Grouped_Data["adopted_user"][i].replace(str(i),"yes")
  else:
    Grouped_Data["adopted_user"][i]=Grouped_Data["adopted_user"][i].replace(str(i),"no")
```

```python
Grouped_Data
```

```python
Grouped_Data["adopted_user"].unique()
```

```python
Grouped_Data["adopted_user"]=Grouped_Data["adopted_user"].replace(["38722","38723"],value="no")
```

```python
Grouped_Data["adopted_user"].unique()
```

```python
Final_data=pd.merge(Grouped_Data, final_data, on='user_id')
```

```python
Final_data
```

```python
Final_data.drop(columns=["time_stamp_y","visited_y"],inplace=True)
```

```python
Final_data.isna().sum()
```

```python
Final_data.info()
```

```python
Final_data["invited_by_user_id"]=Final_data["invited_by_user_id"].astype("int64")
```
